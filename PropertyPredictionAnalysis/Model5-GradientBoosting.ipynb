```python
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor
```

```python
X_train = pd.read_csv('data/X_train.csv')
y_train = pd.read_csv('data/y_train.csv')
X_test = pd.read_csv('data/X_test.csv')
y_test = pd.read_csv('data/y_test.csv')
```

```python
param_grid = {    
    'learning_rate': [0.05, 0.1, 0.25, 0.5, 0.75, 1],
     'max_depth' : np.arange(1,10),
     'n_estimators' : [1,100,1000]}
gb = GradientBoostingRegressor(random_state=42)

gb_cv= GridSearchCV(gb,param_grid,cv=5)
gb_cv.fit(X_train,y_train)
```

```python
gb_cv.best_params_
```

```python
gb = GradientBoostingRegressor(n_estimators = 1000, learning_rate=0.05, max_depth=7, random_state=42)
gb.fit(X_train,y_train)
y_pred=gb.predict(X_test)
r2_score = r2_score(y_test, y_pred)
print("R2 score: %f" % (r2_score))

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE: %f" % (rmse))
```

This is the best so far compared to all models we tried.
